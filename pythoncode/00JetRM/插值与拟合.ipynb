# 插值

```python
# -*- coding: utf-8 -*-
"""
Created on Thu Jul 27 16:53:21 2017
@author: Dell
"""
 
import numpy as np
from scipy import interpolate
import pylab as pl
#创建数据点集并绘制


np.random.seed(10)

pl.figure(figsize=(5,3))
x = np.linspace(0, 20, 21)
y = np.sin(x)+np.random.rand(21)*0.9
ax=pl.plot()
 
pl.plot(x,y,'ro')
#建立插值数据点
xnew = np.linspace(0, 20, 101)
#for kind in ['nearest', 'zero','linear','quadratic']:
for kind in ['nearest','quadratic','cubic',5]:
    #根据kind创建插值对象interp1d
    f = interpolate.interp1d(x, y, kind = kind)
    ynew = f(xnew)#计算插值结果
    pl.plot(xnew, ynew, label = str(kind))
 
pl.xticks(fontsize=20)
pl.yticks(fontsize=20)
 
pl.legend(loc = 2,bbox_to_anchor=(0.99, 0.99))
pl.show()
```

## scipy中 插值函数splrep与interp1d比较


```python

import numpy as np
from scipy.interpolate import interp1d, splrep, splev
 
x = np.linspace(0, 6, 7)
y = np.array([3, 1, 4, 1, 5, 5, 2])    # some data
xx = np.linspace(0, 6, 100)            # evaluation points  
 
y1 = interp1d(x, y, kind='cubic')(xx)
y2 = splev(xx, splrep(x, y, k=3))
print(np.abs(y1-y2).max())

pl.plot(x, y, )
pl.plot(xx, y1, label='interp1d cubic ' )
pl.plot(xx, y2, label='splev k=3 ')

pl.xticks(fontsize=20)
pl.yticks(fontsize=20)
 
pl.legend(loc = 2,bbox_to_anchor=(0.99, 0.99))
pl.show()



y1 = interp1d(x, y, kind='quadratic')(xx)
y2 = splev(xx, splrep(x, y, k=2))
print(np.abs(y1-y2).max())

pl.plot(x, y, )
pl.plot(xx, y1, label='interp1d quadratic ' )
pl.scatter(xx, y2, label='splev k=2 ')

pl.xticks(fontsize=20)
pl.yticks(fontsize=20)
 
pl.legend(loc = 2,bbox_to_anchor=(0.99, 0.99))
pl.show()
```

# 拟合

```python
from scipy.optimize import curve_fit
import numpy as np


def func(x, a, b, c):  # 用来拟合的方程
    return a * np.exp(-b * x) + c


def get_xy():
    xdata: np.ndarray = np.linspace(0, 4, 50)  # x值
    y = func(xdata, 2.5, 1.3, 0.5)
    rng = np.random.default_rng()
    y_noise = 0.2 * rng.normal(size=xdata.size)
    ydata: np.ndarray = y + y_noise  # 拟合的数据 y
    return xdata, ydata



x_value, y_value = get_xy()
popt, pcov = curve_fit(func, x_value, y_value)  # 拟合分布
y_pred = func(x_value, *popt)  # 预测结果


plt.scatter(x_value, y_value, color='b', label='data')
plt.plot(x_value, func(x_value, *popt), 'r-',
         label='fit: a=%5.3f, b=%5.3f, c=%5.3f' % tuple(popt))
# 这个写法是加一个限制：0 <= a <= 3, 0 <= b <= 2 and 0.1 <= c <= 1
popt_2, pcov_2 = curve_fit(func, x_value, y_value, bounds=([0, 0, 0.1], [3., 2., 1]))
plt.plot(x_value, func(x_value, *popt_2), 'g--',
         label='fit: a=%5.3f, b=%5.3f, c=%5.3f' % tuple(popt_2))
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
'''————————————————
版权声明：本文为CSDN博主「呆萌的代Ma」的原创文章，遵循CC 4.0 BY-SA版权协议，转载请附上原文出处链接及本声明。
原文链接：https://blog.csdn.net/weixin_35757704/article/details/121684947'''


```
