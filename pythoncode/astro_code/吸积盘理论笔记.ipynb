dust torus cooling timescale once the active galactic nucleus (AGN) is quenched
In a clumpy torus system, once the incoming photons are suppressed, the cooling timescale of one clump from Tdust = 1000 K
to several 10 K is less than 10 years, indicating that the dust torus cooling time is mainly governed by the light
crossing time of the torus from the central engine


We also propose that those “dying” AGNs could be found using the AGN indicators with a different physical scale R
such as 12 μm band luminosity tracing AGN torus (R ∼ 10 pc) and the optical [O III]λ5007 emission line tracing
narrow line regions (R = 102–4 pc).

Optical Depth
Solving for specific intensity Iν in the Radiative Transport Equation

 \begin{align}  {{dI_\nu \over ds}=j_\nu -\alpha _\nu I_\nu } \,\! \end{align} 
for zero emission (jν = 0) gives a solution of the form:

 \begin{align} I_\nu (s)& =I_\nu (0)e^{-n\sigma _\nu s}\\ & =I_\nu (0)e^{-\tau _\nu }\\ \end{align} \,\!
where τν is the optical depth at ν. Optical depth is often computed as:

 \begin{align} \tau _\nu =n\sigma _\nu s=N\sigma _\nu  \,\! \end{align} 
where N, the column density, is in cm − 2 and is the # of extinguishers per unit area. Similarly,

 \begin{align}\tau _\nu =\rho \kappa _\nu s=\Sigma \kappa _\nu  \,\!\end{align} 
where Σ is the mass surface density.

  \begin{align} \tau _\nu \begin{cases} \ll 1 & optically\  thin\\ \gg 1 & optically\   thick\end{cases} \,\! \end{align}
The Mean Free Path is given by: 
 \begin{align}\lambda _{mfp,\nu }=\alpha _\nu ^{-1}=\frac1{n\sigma _ nu} =\frac1{\rho K_\nu }\end{align} . Thus:

 \begin{align} \tau _\nu ={s\over \lambda _{mfp,\nu }} \,\!\end{align} 
That is, the optical depth is the number of mean-free-paths deep a medium is. For Poisson processes, the probability of absorption is given by:

\begin{align} P(n)={e^{-{s\over \lambda _{mfp,\nu }}}\left({s\over \lambda _{mfp,\nu }}\right)^ n\over n!} \,\!\end{align}
Therefore:

\begin{align}  I_\nu (s)=I_\nu (0)e^{-\alpha _\nu s} \,\!\end{align}
In fact, the radiative transport equation can be expressed in terms of optical depth. Dividing by αν and recognizing dτν = dsαν:

 \begin{align} {dI\nu \over d\tau _\nu }& ={j_\nu \over \alpha _\nu }-I_\nu \\ & =S_\nu -I_\nu \\ \end{align} \,\!　
 
where Sν is a “source function”. In general,

\begin{align}  S_\nu \big |_{scattering}\propto \int {I_\nu d\Omega } \,\!\end{align}
There is a formal solution for Iν. Let’s define \tilde I\equiv Ie^{\tau \nu } and \tilde S\equiv Se^{\tau \nu }. Then:

\begin{align}  {d\tilde I\over d\tau _\nu }=\tilde S \,\!
 \tilde I(\tau _\nu )=\tilde I(0)+\int _0^{\tau _\nu }{\tilde Sd\tilde\tau _\nu } \,\!
  {I_\nu (\tau _\nu )=\overbrace{I_\nu (0)e^{-\tau _\nu }}^{atten\  bg\  light} +\overbrace{\int _0^{\tau _\nu }{S_\nu (\tau _\nu ^\prime ) \underbrace{e^{-(\tau _\nu -\tau _\nu ^\prime )}}_{self-absorption} d\tau _\nu ^\prime }}^{glowing\  medium}} \,\!\end{align}
If Sν is constant with τν, then:

\begin{align}  I_\nu (\tau _\nu )=I_\nu (0)e^{-\tau _\nu }+S_\nu (1-e^{-\tau _\nu }) \,\!\end{align}

That second term on the righthand side can be approximated as Sντν for \tau _\nu \ll 1, since self-absorption is negligible. Similarly, for \tau _\nu \gg 1, it may be approximated as Sν. The source function Sν is everything. It has both the absorption and emission coefficients embedded in it.

What is the ratio of the final intensity to the original intensity?

There are several factors which must affect the intensity in obvious ways:

the greater the distance the beam travels, the more light will be removed from it
the greater the density of material  ρ , the more light will be scattered or absorbed
the composition of the material -- some atoms are much more efficient at absorbing light than others
We can put all this together in mathematical form like so:

dI=−κρIds(28.1)
where  I  is the incoming intensity of the light,  dI  is the amount of light added to the beam (hence the negative sign), ds is the distance it travels, and ρ is the density of the material. The variable  κ , called the opacity or absorption coefficient, serves two purposes: first, it's a constant of proportionality which serves to make the units match on both sides of the equation; second, it expresses the efficiency with which this particular sort of material (hydrogen, or helium, or electrons, or ...) absorbs and scatters light.
