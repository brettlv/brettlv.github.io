```python
import pandas as pd
import numpy as np
import os

def get_obsids(path):
    dirname=os.listdir(path)
    obsids=[]
    for i in dirname:
        if i.isdigit():
            obsids.append(i)
    obsids.sort()        
    return obsids

def drop_index(data):
    data=data.reset_index(drop=True)
    return data

def get_info(data,label,label_err=None):
    return min(data[label]),max(data[label]),np.mean(data[label])
```

## def combine_value_lower_upper

```python
def combine_value_lower_upper(row):

    if row.strip()=='-':
        value,lower,upper=0,0,0
        
    elif row.find('*')>0:
        value,lower,upper=float(row.split('*')[0]),0,0

    elif row.find('\dagger')>0:
        value,lower,upper=float(row.split('$\\dagger$')[0]),0,0
        
    else:    
        value= float(row.split('$')[0].split('{')[-1])
        lower= float(row.split('$_{-')[1].split('}')[0])
        upper= float(row.split('}^{')[1].split('}')[0])
      
    return value,lower,upper
```

```python
logLedd=np.log10(1.263*10**38)+np.log10(Mbh_GRS1915)
```

```python
figure_n= 1
fig = plt.figure(figsize=(6,figure_n*4))
fig.subplots_adjust(hspace=0.0, wspace = 0.0)
ax = fig.add_subplot(figure_n,1,1)


for index,i in enumerate(GRS1915_table5['TBabs']):
    #if i.find('+')>0:
    tbabs_value,tbabs_lower,tbabs_upper=combine_value_lower_upper(i)
        
    #if GRS1915_table5.loc[index,'TBpcf'].find('+')>0:
    TBpcf_value,TBpcf_lower,TBpcf_upper= combine_value_lower_upper(GRS1915_table5.loc[index,'TBpcf'])
            
    #if GRS1915_table5.loc[index,'Fnth(1e-9)'].find('+')>0: 
    Fnth_value,Fnth_lower,Fnth_upper= combine_value_lower_upper(GRS1915_table5.loc[index,'Fnth(1e-9)'])
    color='black'   
    asymmetric_xerror= [[Fnth_lower],[Fnth_upper]]
    asymmetric_yerror= [[TBpcf_lower],[TBpcf_upper]]
        
    ax.errorbar(
                #x=Fnth_value,
                x= np.log10(Fnth_value*lum_const)-logLedd,
                
                #xerr= asymmetric_xerror,
                
                y=TBpcf_value,
                yerr= asymmetric_yerror,
               
               marker='o',ms=11., mew=1, capsize=0,mec=color,ecolor=color,
               elinewidth=2,fmt='o',ls='',fillstyle='none',label='$GRS1915$')
               
#ax.set_xlabel('Fnth(1e-9 erg s$^{-1}$ cm$^{-2}$)')
ax.set_xlabel('log LX/Ledd')
ax.set_ylabel('NH_TBpcf (10$^{22} $cm$^{-2}$)')        
ax.semilogy()
#ax.semilogx()

plt.title('re-brightening phases from NICER observations')
set_ax_legend(ax,bbox_to_anchor=(0.71, 0.99))
#save_lc_img_path='/Users/lyubing/path/lc_%d_%d.png'%(range_l,range_r)
#plt.savefig(save_lc_img_path,dpi=400, transparent=False, bbox_inches='tight')
            
```
