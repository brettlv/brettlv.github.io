```python
from lxml.html import parse
from urllib.request import urlopen
from pandas.io.parsers import TextParser
from pandas import Series
import requests
import pandas as pd
```

```python
def _unpack(row,kind='td'):
    elts=row.findall('.//%s' %kind)
    return [val.text_content() for val in elts]
```

```python
def parse_options_data(table):
    rows=table.findall('.//tr')
    header=_unpack(rows[0],kind='th')
    data=[_unpack(r) for r in rows[1:]]
    #return data
    return TextParser(data,names=header).get_chunk()
```

```python
parsedurl=parse(urlopen(url='http://hxmt.org/index.php/plan/splan'))
doc=parsedurl.getroot()
links=doc.findall('.//a')
for i in links :
    print(i.get('href') ) 
```

```python
import numpy as np 
urllist=[]
for i_ in np.arange(1,8):
    num=4*i_
    print(num)
    parsedurl=parse(urlopen(url='http://hxmt.org/index.php/plan/splan?start=%d'%num))
    doc=parsedurl.getroot()
    links=doc.findall('.//a')
    for i in links :
        if str(i.get('href')).find('hxmt-short-term-schedule')>=0:
            print(i.get('href') )
            urllist.append('http://hxmt.org'+str(i.get('href')))    
```

```python
news_lists = []
for id in urllist:
    if id not in news_lists:
        news_lists.append(id)
```

```python
for url in news_lists[:]:
    print(url)
    df = pd.read_html(url, encoding='utf-8', header=None)[0]
    #print(df.columns)
    data=(df[df[df.columns[3]]=='Point'])
    data.to_csv('point_20190821.csv',mode='a')
```

```python
pointdata=pd.read_csv('point_20190821.csv',encoding='ISO-8859-1')
```

```python
news_name_lists = []
posras=[]
posdecs=[]
for name_,posra,posdec in zip(pointdata['Name'],pointdata['J2000_RA'],pointdata['J2000_Dec']):
    if name_ not in news_name_lists:
        if not name_.startswith('BLANK'):
            news_name_lists.append(name_)
            posras.append(posra)
            posdecs.append(posdec)        
```

```python
news_name_lists,posras,posdecs
```

```python
head = ["name" , "ra" , "dec"]
namelist = [news_name_lists,posras,posdecs]
df = pd.DataFrame(namelist , index = head).T
df.to_csv("pointnames_20190821.csv" , encoding = "utf-8")
```
