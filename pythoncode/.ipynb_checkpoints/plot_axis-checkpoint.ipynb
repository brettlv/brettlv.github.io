```python
from pylab import *
from matplotlib.ticker import MultipleLocator, FormatStrFormatter
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import pyfits


from matplotlib.ticker import NullFormatter  # useful for `logit` scale


from astropy.stats import LombScargle
plt.style.use('seaborn-whitegrid')

%matplotlib inline
%config InlineBackend.figure_format='svg'
#%config InlineBackend.figure_format='retina'


```

```python
xmajorLocator   = MultipleLocator(20) #x10
xmajorFormatter = FormatStrFormatter('%1.1f') #x
xminorLocator   = MultipleLocator(1) #x1
 
ymajorLocator   = MultipleLocator(0.1) #y10
ymajorFormatter = FormatStrFormatter('%2.1f') #y
yminorLocator   = MultipleLocator(0.05) #y1
 
t = np.arange(0.0, 100.0, 1)
s = np.sin(0.1*pi*t)*exp(-t*0.01)
 
ax = plt.subplot(111) #:ax,plot
plt.plot(t,s,'--b*')
 
#,
ax.xaxis.set_major_locator(xmajorLocator)
ax.xaxis.set_major_formatter(xmajorFormatter)
 
ax.yaxis.set_major_locator(ymajorLocator)
ax.yaxis.set_major_formatter(ymajorFormatter)
 
#,
ax.xaxis.set_minor_locator(xminorLocator)
ax.yaxis.set_minor_locator(yminorLocator)
 
ax.xaxis.grid(True, which='major') #x
ax.xaxis.grid(True, which='minor') 
ax.yaxis.grid(True, which='minor') #y
 
show()
```
