As a radio telescope, the VLA observes wavelengths ranging from 7 mm to 4 m, which correspond to microwaves and short wave radio, respectively. (NRAO, 2008) Depending on both the wavelength being observed and the configuration of the array, the array’s field of view and resolution vary as well. With the tightest arrangement of the antennae and the shortest wavelength, the VLA has a FOV of 1.3 arcseconds. In the loosest arrangement and at the longest wavelength, the FOV increases to 5.6 degrees. Similarly, in the tightest arrangement and at the shortest wavelength, the array has an angular resolution of 0.05 arcseconds whereas in the loosest arrangement and at the longest wavelength, the resolution is 14.2 arcminutes. (Chandler, 2009)
