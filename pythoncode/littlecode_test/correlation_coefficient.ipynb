scipy.stats.pearsonr(x, y)
The Pearson correlation coefficient measures the linear relationship between two datasets ã€Œè¡¡é‡ä¸¤ç»„æ•°æ®çš„çº¿æ€§ç›¸å…³æ€§ã€.

The calculation of the p-value relies on the assumption that each dataset is normally distributedã€Œå‡è®¾ä¸¤ç»„æ•°æ®æœä»æ­£æ€åˆ†å¸ƒï¼Œå³æ•°æ®å¿…é¡»æ˜¯è¿ç»­å‹æ•°æ®(continuous)ã€.

Like other correlation coefficients, this one varies between -1 and +1 with 0 implying no correlation. Correlations of -1 or +1 imply an exact linear relationship. Positive correlations imply that as x increases, so does y. Negative correlations imply that as x increases, y decreases.ã€Œpearsonç›¸å…³ç³»æ•°èŒƒå›´ä¸º-1åˆ°1ã€è´Ÿå€¼ä¸ºè´Ÿç›¸å…³ã€0ä¸ºä¸ç›¸å…³ã€æ­£å€¼ä¸ºæ­£ç›¸å…³ä¸‹å›¾èƒ½å¾ˆå¥½çš„å±•ç¤ºè¿™ç§å…³ç³»ğŸ‘‡ğŸ‘‡ã€

```python
from scipy import stats
import numpy as np
a = np.array([0, 0, 0, 1, 1, 1, 1])
b = np.arange(7)
c,p = stats.pearsonr(a, b)
print('Pearson correlationï¼š%s'%c)
print('p-valueï¼š%s'%p)
```

```python
3ã€æ–¯çš®å°”æ›¼ç›¸å…³ç³»æ•°(Spearman)
ä½¿ç”¨å‰æï¼šçš®å°”é€ŠPearsonç›¸å…³ç³»æ•°ä½¿ç”¨å‰ææ¡ä»¶ä¸­ï¼Œä»»ä½•ä¸€ä¸ªæ¡ä»¶ä¸æ»¡è¶³æ—¶å¯ä»¥è€ƒè™‘ä½¿ç”¨è¯¥ç³»æ•°ï¼›

Spearmanä¸Pearsonç›¸å…³ç³»æ•°è®¡ç®—å¾ˆç±»ä¼¼ï¼Œåªæ˜¯Spearmanè®¡ç®—éœ€è¦å°†ä¸¤ä¸ªå˜é‡è½¬åŒ–ä¸ºåºæ•°ï¼Œä»¥ä¸‹ä¸ºscipyä¸­æè¿°ï¼Œ
```

scipy.stats.spearmanr(a, b=None, axis=0, nan_policy='propagate')
Calculate a Spearman correlation coefficient with associated p-value.The Spearman rank-order correlation coefficient is a nonparametric measure of the monotonicity of the relationship between two datasetsã€Œä¸¤ä¸ªå˜é‡æˆå¯¹å–å€¼å¹¶æ’åºå–ç§©ã€. Unlike the Pearson correlation, the Spearman correlation does not assume that both datasets are normally distributedã€Œå‡è®¾ä¸¤ç»„æ•°æ®ä¸éœ€è¦æœä»æ­£æ€åˆ†å¸ƒã€.

Like other correlation coefficients, this one varies between -1 and +1 with 0 implying no correlation. Correlations of -1 or +1 imply an exact monotonic relationship. Positive correlations imply that as x increases, so does y. Negative correlations imply that as x increases, y decreasesã€Œç›¸å…³ç³»æ•°èŒƒå›´ä¸º-1åˆ°1ã€è´Ÿå€¼ä¸ºè´Ÿç›¸å…³ã€0ä¸ºä¸ç›¸å…³ã€.

The p-value roughly indicates the probability of an uncorrelated system producing datasets that have a Spearman correlation at least as extreme as the one computed from these datasets. The p-values are not entirely reliable but are probably reasonable for datasets larger than 500 or soã€Œæ•°æ®é›†å…ƒç´ å¤§å°å¤§äº500å¯èƒ½æ‰é è°±ã€.

```python
#ä½¿ç”¨Python scipy
from scipy import stats
s,p1 = stats.spearmanr([1,2,3,4,5], [5,6,7,8,7])
print('Spearman correlationï¼š%s'%s)
print('p-valueï¼š%s'%p1)
```

scipy.stats.kendalltau(x, y, initial_lexsort=None, nan_policy='propagate', method='auto')

Calculate Kendallâ€™s tau, a correlation measure for ordinal dataã€Œè¯„ä¼°æœ‰åºåˆ†ç±»å˜é‡(ordinal data)çš„ç›¸å…³æ€§ã€.Kendallâ€™s tau is a measure of the correspondence between two rankingsã€Œè¡¡é‡ä¸¤ç»„å˜é‡çš„ç­‰çº§ç›¸å…³æ€§ã€. Values close to 1 indicate strong agreement, values close to -1 indicate strong disagreementã€Œç›¸å…³ç³»æ•°1ä¸ºæåº¦ç›¸å…³ã€-1æåº¦ä¸ç›¸å…³ã€.

```python
from scipy import stats
x1 = [12, 2, 1, 12, 2]
x2 = [1, 4, 7, 1, 0]
k, p2 = stats.kendalltau(x1, x2)
print('Kendallta correlatioï¼š%s'%k)
print('p-valueï¼š%s'%p2)
```
