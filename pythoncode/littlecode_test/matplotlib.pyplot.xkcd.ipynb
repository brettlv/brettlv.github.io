```python
import matplotlib.pyplot as plt
with plt.xkcd(
        scale=4,  #xkcd
        length=120,  #
        randomness=2):  #
    plt.figure(dpi=150)
    patches, texts, autotexts = plt.pie(
        x=[1, 2, 3],  #
        labels=['A', 'B', 'C'],
        colors=['#dc2624', '#2b4750', '#45a0a2'],
        autopct='%.2f%%',
        explode=(0.1, 0, 0))
    texts[1].set_size('20')  #B

    #matplotlib.patches.Wedge
    patches[0].set_alpha(0.3)  #A
    patches[2].set_hatch('|')  #C
    patches[1].set_hatch('x')

    plt.legend(
        patches,
        ['A', 'B', 'C'],  #
        title="Pie Learning",
        loc="center left",
        fontsize=15,
        bbox_to_anchor=(1, 0, 0.5, 1))

    plt.title('Lovely pie', size=20)
    plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
with plt.xkcd():
    plt.figure(dpi=150)
    labels = ['Jack', 'Rose', 'Jimmy']
    year_2019 = np.arange(1, 4)
    year_2020 = np.arange(1, 4) + 1
    bar_width = 0.4

    plt.bar(
        np.arange(len(labels)) - bar_width / 2,  #
        year_2019,
        color='#dc2624',
        width=bar_width,
        label='year_2019'  #
    )
    plt.bar(
        np.arange(len(labels)) + bar_width / 2,
        year_2020,
        color='#45a0a2',
        width=bar_width,
        label='year_2020'  #
    )
    plt.xticks(np.arange(0, 3, step=1), labels, rotation=45)  #
    plt.legend(loc=2)  #
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
#
from sklearn import datasets

iris = datasets.load_iris()
x, y = iris.data, iris.target
pd_iris = pd.DataFrame(np.hstack((x, y.reshape(150, 1))),
                       columns=[
                           'sepal length(cm)', 'sepal width(cm)',
                           'petal length(cm)', 'petal width(cm)', 'class'
                       ])
with plt.xkcd():

    plt.figure(dpi=150)  #
    #plt.style.use('Solarize_Light2')  #Solarize_Light2
    iris_type = pd_iris['class'].unique()  #class
    iris_name = iris.target_names  #
    colors = ['#dc2624', '#2b4750', '#45a0a2']  #
    markers = ['$\clubsuit$', '.', '+']  #

    for i in range(len(iris_type)):
        plt.scatter(
            pd_iris.loc[pd_iris['class'] == iris_type[i],
                        'sepal length(cm)'],  #x
            pd_iris.loc[pd_iris['class'] == iris_type[i],
                        'sepal width(cm)'],  #y
            s=50,  #marker
            c=colors[i],  #marker
            marker=markers[i],  #marker
            #marker=matplotlib.markers.MarkerStyle(marker = markers[i],fillstyle='full'),#marker
            alpha=0.8,  #marker0-1
            facecolors='r',  #markercc
            edgecolors='none',  #marker
            linewidths=1,  #markeredgecolors
            label=iris_name[i])  #label

    plt.legend(loc='upper right')
```
