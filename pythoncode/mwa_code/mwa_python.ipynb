```python
import urllib#
#import urllib2 for python2
import json
from urllib import request#for python3

# Append the service name to this base URL, e.g. 'con', 'obs', etc.
BASEURL = 'http://ws.mwatelescope.org/'

# Some example query data:
fname = '1095675784_20140925102450_gpubox23_01.fits'
starttime = 1095686136


# Function to call a JSON web service and return a dictionary:

def getmeta(servicetype='metadata', service='obs', params=None):
    """Given a JSON web servicetype ('observation' or 'metadata'), a service name (eg 'obs', find, or 'con')
       and a set of parameters as a Python dictionary, return a Python dictionary containing the result.
    """
    if params:
        data = urllib.parse.urlencode(params)  # Turn the dictionary into a string with encoded 'name=value' pairs
    else:
        data = ''
    # Get the data
    try:
        result = json.load(request.urlopen(BASEURL + servicetype + '/' + service + '?' + data))
    except request.HTTPError as error:
        #print "HTTP error from server: code=%d, response:\n %s" % (error.code, error.read())
        return
    except request.URLError as error:
        #print "URL or network error: %s" % error.reason
        return
    # Return the result dictionary
    return result



```

```python
import os
import numpy as np
import re
obsids=np.loadtxt('CygX-3_zero_list.txt',dtype=np.str,unpack=True,usecols=[0])

def make_cal2_file(obsid):
    with open('job_cal2_mwa_%s.sh'%obsid,'w+') as f:
        f.write('#BSUB -J mwa_brettlv \n')
        f.write('#BSUB -q serial \n')  # (queue name)
        f.write('#BSUB -n 1 \n') #  (use one processor)
        f.write('#BSUB -o %J.out \n')# (result output)
        f.write('#BSUB -e %J.err \n')# (error information)
        f.write('bash mwa_cal2_%s.sh  \n'%obsid) #(executive program name)

def make_beam_file(obsid):
    with open('job_beam_mwa_%s.sh'%obsid,'w+') as f:
        f.write('#BSUB -J mwa_brettlv \n')
        f.write('#BSUB -q serial \n')  # (queue name)
        f.write('#BSUB -n 1 \n') #  (use one processor)
        f.write('#BSUB -o %J.out \n')# (result output)
        f.write('#BSUB -e %J.err \n')# (error information)
        f.write('bash mwa_beam_%s.sh  \n'%obsid) #(executive program name)        
        
        
def make_mwa_file(obsid):
    if not os.path.exists('mwa_cal2_%s.sh'%obsid):
        os.system('cp cal2.sh mwa_cal2_%s.sh'%obsid)
    if not os.path.exists('mwa_beam_%s.sh'%obsid):    
        os.system('cp beam_obsnum.sh mwa_beam_%s.sh'%obsid)
    
    
def replace_obsnum_cal2(obsid):
    f=open('mwa_cal2_%s.sh'%obsid,'r')
    text=f.readlines()
    f.closed        
    with open('mwa_cal2_%s.sh'%obsid,'w+') as f:
        for eachline in text:
            a=re.sub('1205016656',obsid,eachline)
            f.writelines(a)
            
def replace_obsnum_beam(obsid):
    f=open('mwa_beam_%s.sh'%obsid,'r')
    text=f.readlines()
    f.closed        
    with open('mwa_beam_%s.sh'%obsid,'w+') as f:
        for eachline in text:
            a=re.sub('1205016656',obsid,eachline)
            f.writelines(a)            
            
        
def main():
    for obsid in obsids:
        make_cal2_file(obsid)
        make_beam_file(obsid)
        make_mwa_file(obsid)
        replace_obsnum_cal2(obsid)
        replace_obsnum_beam(obsid)
        
        
if __name__ == '__main__':
    main()    
            
```
